```python
from ultralytics import YOLO

from IPython.display import Image, display
```

```python
model = YOLO('yolov8x')
```

```python
results = model.predict('input_videos/match.mp4', save=True)
# save = True allows us to save the resulting video to local sys instead of saving it in the variable

print(results[0])

for box in results[0].boxes:
    print(box)
```
